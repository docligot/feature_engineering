```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
df = pd.read_csv(r"C:\Users\Cirrolytix\Documents\Python Bootcamp\linear_exercise.csv")
```

```python
df
```

```python
df["scaled_leaves"] = preprocessing.scale(df["AvgLeave Balance"])
df["scaled_fte"] = preprocessing.scale(df["FTEs"])
df["scaled_positions"] = preprocessing.scale(df["Open Positions"])
```

```python
df
```

```python
df["leave_per_fte"] = df["AvgLeave Balance"] / df["FTEs"] 
```

```python
df
```

```python
region_dummies = pd.get_dummies(df["Region"])
dept_dummies = pd.get_dummies(df["Department"])
```

```python
df = df.join(region_dummies)
```

```python
df = df.join(region_dummies)
df = df.join(dept_dummies)
```

```python
df
```

```python
pred1 = df[["FTEs", "Open Positions", "AvgLeave Balance"]]
pred2 = df[["scaled_leaves", "scaled_fte", "scaled_positions", "leave_per_fte"]]
pred3 = df[["CALABARZON", "Cebu", "Davao", "Finance", "HR", "IT"]]
pred4 = df[["scaled_leaves", "scaled_fte", "scaled_positions", "leave_per_fte", "CALABARZON", "Cebu", "Davao", "Finance", "HR", "IT"]]
pred5 = df[["FTEs", "Open Positions", "AvgLeave Balance", "CALABARZON", "Cebu", "Davao", "Finance", "HR", "IT"]]
```

```python
pred5
```

```python
lm.fit(pred1, df["Overtime Hrs"])
lm.score(pred1, df["Overtime Hrs"])
```

```python
lm.fit(pred3, df["Overtime Hrs"])
lm.score(pred3, df["Overtime Hrs"])
```

```python
lm.fit(pred2, df["Overtime Hrs"])
lm.score(pred2, df["Overtime Hrs"])
```

```python
lm.fit(pred4, df["Overtime Hrs"])
lm.score(pred4, df["Overtime Hrs"])
```

```python
lm.fit(pred5, df["Overtime Hrs"])
lm.score(pred5, df["Overtime Hrs"])
```
